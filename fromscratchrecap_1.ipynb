```python
dfs["Close"].plot(figsize=(12,6))
dfs.rolling(window=7).mean()["Close"].plot()
```

```python
title = "TITLE"
xlabel = "Date"
ylabel = "Close"
```

```python
ax =  dfs["Close"].plot(figsize=(12,6),title=title)
ax.set(xlabel=xlabel,ylabel=ylabel) 
```

```python
ax = dfs["Close"].plot(figsize=(12,6),title=title)
ax.autoscale(axis="both", tight=True)
ax.set(xlabel=xlabel,ylabel=ylabel)
```
