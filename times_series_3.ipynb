```python
title = "TITLE"
ylabel = "Y LABEL"
xlabel = "X LABEL"
```

```python
df["Close"].plot(figsize=(12,6), title=title)
```

```python
ax = df["Close"].plot(figsize=(12,6),title=title)
ax.set(xlabel=xlabel,ylabel=ylabel)
```

```python
ax = df["Close"].plot(figsize=(12,6),title=title)
ax.autoscale(axis="both", tight=True)
ax.set(xlabel=xlabel,ylabel=ylabel)
```

```python
ax = df["Close"].plot(figsize=(12,6),title=title)
ax.autoscale(axis="y", tight=True)
ax.set(xlabel=xlabel,ylabel=ylabel)
```
