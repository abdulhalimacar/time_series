```python
import pandas as pd
import numpy as np
from statsmodels.tsa.filters.hp_filter import hpfilter
```

```python
df = pd.read_csv("macrodata.csv",index_col=0,parse_dates=True)
df
```

```python
gdp_cycle,gdp_trend = hpfilter(df["realgdp"],lamb=1600)
```

```python
gdp_trend.plot()
```

```python
gdp_cycle.plot()
```

```python
df["trend"]=gdp_trend
```

```python
df[["realgdp","trend"]].plot()
```
